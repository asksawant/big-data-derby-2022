### Classifying the race based on distance

One furlong equals 1/8 of a mile
- 1 mile = 8 furlong
- 7/8 mile = 7 furlong
- 3/4 mile = 6 furlong and so on

Classifying the race based on distance
- All races UNDER one mile (8 furlongs) are referred to as SPRINTS.
- All races one mile (8 furlongs) and OVER are referred to as ROUTES.
- So we can classify the race based on spirt and routes

```python
rdf.loc[rdf['distance_id'] < 800,'rt_furlong'] = 'sprint'
rdf.loc[rdf['distance_id'] >= 800,'rt_furlong'] = 'route'
```
